**Evaluacion y segmentación entre las 09:00 y las 12:00**

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')

# rango horario (, de 09:00 a 12:00)
rango_inicio = pd.to_datetime('09:00', format='%H:%M').time()
rango_fin = pd.to_datetime('12:00', format='%H:%M').time()

# Filtra el DataFrame por el rango horario
df_filtrado = df[(df['Time'].dt.time >= rango_inicio) & (df['Time'].dt.time <= rango_fin)]

# Ahora df_filtrado contiene solo las filas con ventas que ocurrieron entre las 09:00 y las 12:00
```

```python
print(df_filtrado.shape)
```

```python
# contar los valores únicos en la columna 'Gender'
conteo_tipos = df_filtrado['Gender'].value_counts()

# Imprime el conteo de cada tipo
print(conteo_tipos)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(conteo_tipos, labels=conteo_tipos.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Géneros entre compradores de 9 a 12')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el conteo de cada tipo
conteo_tipos = male_data['Payment'].value_counts()

# Definir el tamaño del gráfico
plt.figure(figsize=(8, 6))

# Crear el gráfico de barras
sns.barplot(x=conteo_tipos.index, y=conteo_tipos.values, palette='pastel')

# Personalizar el gráfico
plt.title('Métodos de Pago para Male Customers')
plt.xlabel('Método de Pago')
plt.ylabel('Frecuencia')

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```
